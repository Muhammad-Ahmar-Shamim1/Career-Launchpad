```python
# src/data_processing.py

import pandas as pd
import os

def preprocess_data(input_path, output_path):
    df = pd.read_csv(input_path)

    # Rename columns
    df.rename(columns={
        'Date': 'date',
        'Units Sold': 'sales',
        'Inventory Level': 'inventory_level',
        'Competitor Pricing': 'competitor_price',
        'Holiday/Promotion': 'holiday'
    }, inplace=True)

    # Convert date
    df['date'] = pd.to_datetime(df['date'])

    # Sort by date
    df = df.sort_values('date')

    # Handle missing values
    df['sales'] = df['sales'].fillna(0)
    df['Price'] = df['Price'].fillna(df['Price'].mean())
    df['Discount'] = df['Discount'].fillna(0)
    df['inventory_level'] = df['inventory_level'].ffill()

    # ðŸ”¹ CREATE DIRECTORY IF NOT EXISTS
    os.makedirs(os.path.dirname(output_path), exist_ok=True)

    # Save processed data
    df.to_csv(output_path, index=False)
    print("âœ… Data Processing Completed Successfully")

if __name__ == "__main__":
    preprocess_data(
        input_path=r"F:\Career Launchpad\sales_forecasting week-1\data\retail_store_inventory.csv",
        output_path=r"F:\Career Launchpad\sales_forecasting week-1\data\processed\sales_processed.csv"
    )
```
